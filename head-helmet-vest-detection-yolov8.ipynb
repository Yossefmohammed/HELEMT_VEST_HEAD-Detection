```python
from ultralytics import YOLO
model = YOLO('yolov8l.pt')  # ✅ correct


# Train the model
model.train(
    data='data.yaml',         # Path to your YAML file
    epochs=15,                # Number of epochs
    imgsz=640,                # Image size
    batch=16,                 # Batch size
    device=0                  # Use GPU (0 for first GPU)
)
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import glob

# Load the model
model = YOLO('runs/detect/train/weights/best.pt')

# Get list of 5 test images (adjust the path as needed)
image_paths = glob.glob('/kaggle/input/hardhat-vest-dataset-v3/images/test/*.jpg')[:15]

# Loop through images and predict
for path in image_paths:
    results = model(path)  # Run detection
    detected_image = results[0].plot()  # Get annotated image

    # Display the image
    plt.figure(figsize=(8, 8))
    plt.imshow(cv2.cvtColor(detected_image, cv2.COLOR_BGR2RGB))
    plt.title(f"Detection: {path.split('/')[-1]}")
    plt.axis('off')
    plt.show()
```
